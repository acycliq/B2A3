## How to make cellData_rgb.tsv
### The original name of his notebook was cellData_rgb.ipynb and was saved at /media/dimitris/New Volume/data/Izzie/221020_IF_primary_then_seq/pciSeq

```python
import pandas as pd
import numpy as np
import json
from PIL import ImageColor
```

```python
cellData = pd.read_csv(r"E:\data\Izzie\221020_IF_primary_then_seq\pciSeq\cellData.tsv", sep='\t')

cellData['Genenames'] = cellData['Genenames'].apply(lambda x: eval(x))
cellData['CellGeneCount'] = cellData['CellGeneCount'].apply(lambda x: eval(x))
cellData['ClassName'] = cellData['ClassName'].apply(lambda x: eval(x))
cellData['Prob'] = cellData['Prob'].apply(lambda x: eval(x))
cellData['ellipsoid_border'] = cellData['ellipsoid_border'].apply(lambda x: eval(x))
cellData['sphere_scale'] = cellData['sphere_scale'].apply(lambda x: eval(x))
cellData['sphere_rotation'] = cellData['sphere_rotation'].apply(lambda x: eval(x))
```

```python
cellData.head()
```

```python
cellData['BestClass'] = cellData.apply(lambda r: r['ClassName'][np.argmax(r['Prob'])], axis=1)
cellData['BestProb'] = cellData.apply(lambda r: np.max(r['Prob']), axis=1)
```

```python
cellData = cellData[['Cell_Num', 'X','Y','Z','ClassName','Prob','sphere_scale', 'sphere_rotation', 'BestClass']] 
cellData.head()
```

```python
f = open(r'E:\data\Izzie\221020_IF_primary_then_seq\pciSeq\classColours.json')
data = json.load(f)
```

```python
classColours = pd.DataFrame(data)
classColours.sort_values('className')
```

```python
uname, id = np.unique(classColours.IdentifiedType, return_inverse=True)
```

```python
classColours['class_id'] = id
classColours
```

```python
# use this to make the json for the classCOnfig.js file. You migh have to manually 
# change strings like "class_id":0 to "class_id":"class_0" and maybe use this as the className key
classColours.to_json('classConfig2.js', orient='records')
```

```python
r,g,b = list(zip(*classColours.color.apply(lambda d: ImageColor.getcolor(d, "RGB")).values))
```

```python
classColours['r'] = r
classColours['g'] = g
classColours['b'] = b
classColours
```

```python
uname, id = np.unique(classColours.className, return_inverse=True)
```

```python
mydict = {}
mydict.update(zip(classColours.className, classColours.class_id))
```

```python
classes = cellData.ClassName.apply(lambda d: [mydict[v] for v in d])
```

```python
cellData['classes'] = classes
cellData.shape
```

```python
cellData = cellData.merge(classColours, left_on='BestClass', right_on='className')
cellData.shape
```

```python
cellData = cellData[['Cell_Num', 'X','Y','Z','classes','Prob', 'sphere_scale', 'sphere_rotation', 'r', 'g', 'b']] 
```

```python
cellData = cellData.rename(columns={"Cell_Num": "label", 
                                    "X": "x",
                                    "Y": "y",
                                    "Z": "z",
                                    "Prob": "class_prob"})
```

```python
cellData = cellData[["label", "x", "y", "z", "classes", "class_prob", "sphere_scale", "sphere_rotation", "r", "g", "b"]]
cellData.x = cellData.x - 3893.48 # substract the spots.x.mean()
cellData.y = cellData.y - 4150.75 # substract the spots.y.mean()
cellData.z = cellData.z - 96.79   # substract the spots.z.mean() 
```

```python
cellData.to_csv("cellData_rgb.tsv", sep='\t', index=False)
```
